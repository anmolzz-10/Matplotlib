```python
import matplotlib.pyplot as plt
```

### Matplotlib Functions
##### plt.plot(x,y,color= 'color name',linestyle= 'line_style', linewidth = value ,marker = 'marker symbol', label = 'label name')

```python
months = [1,2,3,4]
sales = [1000,1500,1200,1800]
plt.plot(months,sales, color = 'blue', linestyle = '--', linewidth = 2,marker= 'o', label ='2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales data report')
plt.legend(loc= 'upper left', fontsize = 12)
plt.grid(color= 'gray', linestyle = ':',linewidth= 1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4], ['M1','M2','M3','M4']) ##replace actual data points with relevant values like Month in this case 
plt.show()
```

### Matplotlib Graphs

#####  1.) Bar Chart Used for Categorical plt.bar(category,value,color = 'colorname', width =value,label='labelname')

```python
x =['A','B','C','D']
y= [1000,1500,800,1200]
plt.bar(x,y,color ='orange',label ='Sales 2025') ##use plt.barh for horizontal bar
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend()
plt.show()
```

##### 2.) Pie Chart plt.pie(values,label= label_list, colors= color_list,autopct ='%1.1f%%')

```python
regions =['North','South','West','East']
revenue =[3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct ='%1.1f%%',colors= ['gold','skyblue','lightgreen','coral'])
plt.title('Revenue Contribution by region')
plt.show()
```

### 3.) Histogram plt.hist(data,bins=number_of_bins,color='colorname',edgecolor ='black')

```python
scores =[45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]
plt.hist(scores,bins=5,color='purple',edgecolor ='black')
plt.xlabel('Score Range') ##data ranges
plt.ylabel('Number of Students') ##frequency of the data 
plt.title('Distribution of students')
plt.show()
```

####  4.) Scatter Plots plt.scatter(x,y,color ='color_name, marker ='marker style',label ='label name')

```python
x = [1,2,3,4,5,6,7,8]
y = [50,55,60,65,70,75,80,85]
plt.scatter(x,y,color ='red',marker ='^',label = 'Student Data')
plt.xlabel('Hours studied')
plt.ylabel('Exam Score')
plt.title('Relationship between hours studied and marks')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.scatter([1,2,3],[50,55,60],color = 'blue',label = 'Class A')
plt.scatter([1,2,3],[45,50,62],color = 'orange',label = 'Class B')
plt.xlabel('hours studied')
plt.ylabel('exam score')
plt.legend()
plt.grid(True)
plt.show()
```

#### Sub Plots used in EDA

#### plt.subplot(nrows,ncols,index)

```python
import matplotlib.pyplot as plt
x =[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1)## 1st row 2nd clumn 1st subplot 
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) ## 1st row 2nd clumn 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

#fig ax = plt.subplots(nrows,ncols, figsize =(width,height))

fig,ax = plt.subplots(1,2, figsize= (10,5))
x =[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y, color  = 'blue')
ax[0].set_title('Line Chart')

ax[1].bar(x,y, color = 'green')
ax[1].set_title('Bar Chart')


fig.suptitle('Comparison of Line and Bar charts')


plt.tight_layout()
plt.show()
```
